#### Structure of the dataset. Reference. 

This dataset includes a list of video games that have sold more than 100,000 copies. It was created by scraping vgchartz.com.
There are 16,598 records.

Rank - ranking of overall sales

Name - name of the game

Platform - platform of the games release (Xbox360, PC, PS4, etc.)

Year - year when the game was released

Genre - game genre

Publisher - publisher of the game

NA_Sales - sales in North America (in millions)

EU_Sales - sales in Europe (in millions)

JP_Sales - sales in Japan (in millions)

Other_Sales - sales in other parts of the world (in millions)

Global_Sales - worldwide sales.

#### Source: https://www.kaggle.com/datasets/gregorut/videogamesales

```r
year_release_frequency <- data.frame(cbind(frequency = table(vg_data$Year), 
                                           percent = round(prop.table(table(vg_data$Year)) * 100,2)))

year_release_frequency <- year_release_frequency[order(year_release_frequency$frequency, decreasing=TRUE), ]
year_release_frequency  %>% head
```

```r
freq_plt <- ggplot(year_release_frequency, aes(x = frequency, y = row.names(year_release_frequency)))+
geom_bar(stat = 'identity', fill="#FF9999")+
# geom_label(mapping = aes(label=percent), size = 3, fontface = "bold", hjust=0.0001)+
labs(title = "Game releases frequency.", x = 'Frequency',y = 'Year')

# freq_plt 
ggplotly(freq_plt, width = 950, height = 950)
```

From this graph, it can be seen that starting from 1980 the number of games released were below 100 games until 1993. In 1994 publishers began to release more games. According to this dataset, in 2009 publishers released the highest number of games, reaching 1431 releases. Previous year was almost the same, having a number of 1428 games. However, the frequency of the releases started decreasing in 2010 from 1259 to 344 in 2016. Moreover, it can be seen, that there is almost no information about game releases in 2017 and 2020.

```r
genre_example_set <- count(vg_data, game_genre=Genre)
genre_example_set$percent <- paste(round((genre_example_set$n/sum(genre_example_set$n))*100,2),'%') # round( number * 100, round to 2 decimal places)

genre_example_set <- genre_example_set[order(genre_example_set$n, decreasing=TRUE), ]
genre_example_set
```

```r
popular_genre_plt <- genre_example_set %>% 
ggplot(aes(x=game_genre,y=n,fill=game_genre))+
geom_bar(stat='identity')+
coord_flip()+
labs(title = "Game releases according to their genres", x = "Genre", y= "Releases")

ggplotly(popular_genre_plt)
```

* It is clearly seen that Action is the most popular genre having 3253 (19.92%) games published. Whereas, The second most popular genre is Sports, which reached 2304 (14.11%) games, followed by Misc and Role-Playing genres, with 1710 (10.47%) and 1471 (9.01%) games respectively. Shooters are on the 5th place, there were 1282 (7.85%) games released.

```r
popular_genres_by_years <- vg_data %>% 
separate_rows(Genre,sep=", ") %>%
filter(Genre %in% genre_example_set$game_genre) %>% 
mutate(Genre=factor(Genre,levels = genre_example_set$game_genre)) %>% 
group_by(Year,Genre) %>% 
summarise(count=n())

popular_genres_by_years 
```

```r
pop_gen_plt <- popular_genres_by_years%>% 
ggplot(aes(x=Year,y=count,fill=Genre))+
geom_bar(stat='identity')+
coord_flip()+
labs(title = "Game releases according to their genres from 1980 to 2020.", x = "Year", y= "Number of releases")

# ggplotly(pop_gen_plt, width = 950, height = 950)
pop_gen_plt
```

* From this graph we can conclude that Action the most popular genre in 1981 and 1982. There were not so many game releases until 1992. 
* Sport games were the most popular genre from 1993 to 2002, however, at the beginning of 2003 this trend was replaced by Action games. They held their leading position until 2016.
* There is too little info about game releases in 2017 and 2020, probaply this dataset was not updated since that period.

```r
filtered_data <- vg_data %>% 
separate_rows(Genre,sep=", ") %>%
filter(Genre %in% genre_example_set$game_genre) %>% 
mutate(Genre=factor(Genre,levels = genre_example_set$game_genre)) %>% 
group_by(Platform,Genre) %>% 
summarise(count=n())

pl_2 <- filtered_data %>% 
ggplot(aes(x=Platform,y=count,fill=Genre))+
geom_bar(stat='identity')+
coord_flip()+
labs(title = "Various game genres published on different platforms", x = "Platforms", y= "Number of releases")

# ggplotly(pl_2,width = 950, height = 950)
pl_2
```

This barchart illustrates the number of games published on different platforms according to their genres. At first glance, the bar chart may look messy, however with plotly library it is possible to obtain precise values of every genre for each of the available platform.

* It can be seen that PS2 and DS are the most popular platforms where games were released
* PS3, Wii and Xbox360 are on the 3rd, 4th and 5th place respectively
* According to the chart it can be seen that Action genre is very popular among all the platforms.
* Misc genre was the most popular for DS and Wii platforms.
* Sports genre is the most frequent genre released on PS2, whereas Strategy was the most popular genre for the PCs among other releases.


#### Finding outliers

```r
vg_data %>% head()
```

```r
sales_sample <- data.frame('year'= c(vg_data$Year),
                           'platform'= c(vg_data$Platform),
                           'genre'= c(vg_data$Genre),
                           'na_sales'= c(vg_data$NA_Sales),
                           'eu_sales'= c(vg_data$EU_Sales),
                           'jp_sales'= c(vg_data$JP_Sales),
                           'other_sales'= c(vg_data$Other_Sales),
                           'global_sales'= c(vg_data$Global_Sales))

sales_sample %>% head
```

```r
na_plt <- ggplot(sales_sample, aes(x=year, y=global_sales, color=year))+
geom_boxplot()+
coord_flip()+
labs(title = 'Global_Sales from 1980 to 2020')

# ggplotly(na_plt, width = 700, height = 700)
na_plt
```

* It can be seen that there many outliers present. They can cause bias if this dataset will be used for ML. So they must be removed before training.

I can use **sigma rule** and try to clean the data.

SIGMA RULE states for removing everything which is higher or lower than 3 SD from the mean

```r
# vg_data <- read.csv("../input/videogamesales/vgsales.csv", stringsAsFactors = FALSE)
library(outliers)

data_outliers <- scores (sales_sample$global_sales) #z score standardization
mean(data_outliers)
mean(sales_sample$global_sales)

sales_sample[data_outliers > 3 | data_outliers < -3, "global_sales"] <- NA
sales_sample[data_outliers > 3 | data_outliers < -3, "na_sales"] <- NA
sales_sample[data_outliers > 3 | data_outliers < -3, "eu_sales"] <- NA
sales_sample[data_outliers > 3 | data_outliers < -3, "jp_sales"] <- NA
sales_sample[data_outliers > 3 | data_outliers < -3, "other_sales"] <- NA
```

#### Correlation matrix

The goal of this analysis to create a correlation matrix and identify what type of sales can affect the most to each other

```r
vg_data %>% head
```

```r

correlation_set <- data.frame('year' = c(vg_data$Year),
                              'na_sales'= c(vg_data$NA_Sales),
                              'eu_sales'= c(vg_data$EU_Sales),
                              'jp_sales'= c(vg_data$JP_Sales),
                              'other_sales'= c(vg_data$Other_Sales),
                              'global_sales'= c(vg_data$Global_Sales))
# correlation_set
```

```r
# install.packages("reshape2")
library(reshape2)
 
# creating correlation matrix
correlation_matrix <- round(cor(correlation_set),2)
correlation_matrix

# reordering corr matrix using corr coefficient as distance metric
dist <- as.dist((1-correlation_matrix)/2)
 
# hierarchical clustering the dist matrix
hc <- hclust(dist)
correlation_matrix <-  correlation_matrix[hc$order, hc$order]
 
# reducing the size of correlation matrix
melted_correlation_matrix <- melt(correlation_matrix)
 
#plotting the correlation heatmap
cormat_plt <- 
library(ggplot2) # comment this line for plotly interractive heatmap
ggplot(data = melted_correlation_matrix, aes(x=Var1, y=Var2, fill=value)) +
geom_tile()+
scale_fill_gradient2(low = 'red', high = 'blue', mid = 'white',midpoint = 0, limit = c(-1,1), space = "Lab", 
   name="Pearson\nCorrelation")

cormat_plt
# ggplotly(cormat_plt, width = 700, height = 700)
```

For some reason the same code is not working in Jupyter notebook. It outputs an error saying that data should be numeric. 
But, infact it is numeric. The same code works perfectly in Kaggle R notebooks. So I will just paste output from there.

* Global Sales are highly correlated with NA Sales, their Pearson Correlation is equal to 0.94.
* EU Sales are correlated with NA Sales and Global Sales having values of 0.77 and 0.90 respectively.
* Other Sales are correlated with EU Sales (0.73), NA Sales (0.63) and Global Sales (0.61)

From this correlation matrix it also can be clearly seen that Year variable is not really correlated with the other variables. Wereas Sales in North America have a significant impact on Global Sales.
